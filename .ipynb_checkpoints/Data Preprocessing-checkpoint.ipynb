## removing outliers

```python
# Calculate the IQR to remove the outliers
Q1 = All_datasets_withoutNaN.quantile(0.25)
Q3 = All_datasets_withoutNaN.quantile(0.75)
IQR = Q3 - Q1
print("IQR is\n", IQR)
print((All_datasets_withoutNaN < (Q1 - 1.5 * IQR)) |(All_datasets_withoutNaN > (Q3 + 1.5 * IQR)) )
print(All_datasets_withoutNaN.shape)
# Applying IQR
df_out = All_datasets_withoutNaN[~((All_datasets_withoutNaN < (Q1 - 1.5 * IQR)) |(All_datasets_withoutNaN > (Q3 + 1.5 * IQR))).any(axis=1)]
df_out
```

## data distribution after removing outliers

```python
data_distribution_after = df_out.hist(bins=100, figsize = (20,15))
```

```python
# convert encoded_label column from float to string to classify the labels and visualize them
df_out['encoded_label'] = df_out['encoded_label'].astype(str)
# filtering sensor data for specific activities
sitting = df_out.query('encoded_label == "4.0"')
standing = df_out.query('encoded_label in ["5.0"]')
walking = df_out.query('encoded_label == "9.0"')
downstairs = df_out.query('encoded_label == "1.0"')
upstairs = df_out.query('encoded_label == "7.0"')
typing = df_out.query('encoded_label == "6.0"')
reading = df_out.query('encoded_label == "3.0"')
hand_wash = df_out.query('encoded_label == "2.0"')
dishes = df_out.query('encoded_label == "0.0"')
vacuum = df_out.query('encoded_label == "8.0"')
```

## apply sliding window and feature extraction

```python
# Sliding window (rolling) with a fixed window size of 1 second for 60Hz
window_size =1*60

#sliding window and feature extraction

# X=AllSubjectsDF.iloc[:,1:]
# X_mean=X.rolling(256).mean()
# X_std=X.rolling(256).std()
# AllSubjectsDF.head(256)

# Features: minimum (min), maximum (max), sum (sum), standard deviation (std), mean (mean)
# selecting datafarame without the labels
xx=df_outWTime.iloc[:,:-1]
features = xx.rolling(window = window_size, center=True, min_periods=1).agg(['min', 'max', 'sum', 'mean', 'std'])
features.columns = ['-'.join(tup).rstrip('-') for tup in features.columns.values]
    # features.dropna(inplace=True)
features
```

```python
# selecting the data and the labels which is the last column
X = df_out.iloc[:, :-1].values
y = df_out.iloc[:, -1].values
# when the label at the first
# x = df_out.iloc[:, 1:].values
# y = df_out.iloc[:, :1].values
```

```python
#splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=1000)
```

```python
# standarization or normalization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:] = sc.fit_transform(x_train[:])
x_test[:] = sc.transform(x_test[:])
```

```python
x_train
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
print('GaussianNBModel Train Score is : ' , classifier.score(x_train, y_train))
print('GaussianNBModel Test Score is : ' , classifier.score(x_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
